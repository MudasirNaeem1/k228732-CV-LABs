*****Task 1 (Grocery List)*****

Your task is to create a simple program to manage a grocery list. Your program should allow the user to add items to the list, remove items, and display the current list of items. Write a Python program that implements this grocery list management system.****

```python
groceries = ['Milk', 'Eggs', 'Bread']

def add(item):
    groceries.append(item)

def remove(item):
    if item in groceries:
        groceries.remove(item)
    else:
        print("Item not found.")

def show():
    print("Current List:", groceries)

while True:
    print("Choose an option:")
    print("1. Show List")
    print("2. Add Item")
    print("3. Remove Item")
    print("4. Exit")

    choice = int(input("Enter (1-4): "))

    if choice == 1:
        show()
    elif choice == 2:
        item = input("Enter item to add: ")
        add(item)
        print("Item has been Added Successfully.")
    elif choice == 3:
        item = input("Enter item to remove: ")
        remove(item)
        print("Item has been Removed Successfully.")
    elif choice == 4:
        print("Program exited.")
        break
    else:
        print("Invalid option, try again.")
```

*****Task 2 (Student Record System)*****

Your task is to build a simple student record system. Each student is identified by their
student ID, and you need to store their names and corresponding grades. Your program should allow adding
students, updating their grades, and displaying the student records. Write a Python program that implements this
student record system using a dictionary.

```python
students = {}

def add_student(sid, name, grade=None):
    students[sid] = {"name": name, "grade": grade}

def update_grade(sid, grade):
    if sid in students:
        students[sid]["grade"] = grade
    else:
        print("Student ID not found.")

def show():
    if not students:
        print("No records yet.")
    else:
        for sid, info in students.items():
            print("ID ", sid, "=", info)

while True:
    print("\nChoose an option:")
    print("1. Show Students")
    print("2. Add Student")
    print("3. Update Grade")
    print("4. Exit")

    choice = int(input("Enter (1-4): "))

    if choice == 1:
        show()
    elif choice == 2:
        sid = input("Enter Student ID: ")
        name = input("Enter Name: ")
        grade = input("Enter Grade: ")
        add_student(sid, name, grade)
        print("Student Added.")
    elif choice == 3:
        sid = input("Enter Student ID: ")
        grade = input("Enter New Grade: ")
        update_grade(sid, grade)
        print("Grade Updated.")
    elif choice == 4:
        print("Program exited.")
        break
    else:
        print("Invalid option, try again.")
```

*****Task 7*****

Load a grayscale image and apply binary thresholding to it and the rotate at 60 degree using OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

print("Upload an image from your PC:")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]

    # Read image in grayscale
    gray = cv2.imread(filename, 0)

    # Apply threshold
    _, th = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # Get image dimensions and create rotation matrix
    h, w = gray.shape
    M = cv2.getRotationMatrix2D((w/2, h/2), 60, 1)
    rot = cv2.warpAffine(gray, M, (w, h))

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(gray, cmap="gray")
    plt.title("Grayscale")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(th, cmap="gray")
    plt.title("Threshold")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(rot, cmap="gray")
    plt.title("Rotated 60°")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    print("Image transformations completed!")
    print(f"Image size: {w}x{h} pixels")
else:
    print("No file uploaded.")
```

*****Task 8*****

Load two images and blend them using OpenCV, and the convert it to grayscale and apply histogram equalization
to enhance contrast.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

print("Upload the first image:")
uploaded1 = files.upload()

if uploaded1:
    filename1 = list(uploaded1.keys())[0]
    a = cv2.imread(filename1)

    # second image
    print("📤 Upload the second image:")
    uploaded2 = files.upload()

    if uploaded2:
        filename2 = list(uploaded2.keys())[0]
        b = cv2.imread(filename2)

        # Resize second image to match first image dimensions
        b = cv2.resize(b, (a.shape[1], a.shape[0]))
        blend = cv2.add(a, b)

        # Convert to grayscale and apply histogram equalization
        g = cv2.cvtColor(blend, cv2.COLOR_BGR2GRAY)
        eq = cv2.equalizeHist(g)

        plt.figure(figsize=(12, 10))

        plt.subplot(2, 2, 1)
        plt.imshow(cv2.cvtColor(a, cv2.COLOR_BGR2RGB))
        plt.title("Image 1")
        plt.axis('off')

        plt.subplot(2, 2, 2)
        plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB))
        plt.title("Image 2 (Resized)")
        plt.axis('off')

        plt.subplot(2, 2, 3)
        plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB))
        plt.title("Blended")
        plt.axis('off')

        plt.subplot(2, 2, 4)
        plt.imshow(eq, cmap="gray")
        plt.title("Histogram Equalized")
        plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("Image blending completed!")
        print(f"Final size: {a.shape[1]}x{a.shape[0]} pixels")
    else:
        print("Second image not uploaded.")
else:
    print("First image not uploaded.")
```

*****Task 9*****

Create binary images and perform bitwise AND, OR, XOR, and NOT operations using OpenCV and NumPy.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

print("Upload an image from your PC:")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]
    img = cv2.imread(filename)
    h, w, _ = img.shape

    # Create rectangular mask
    mask = np.zeros((h, w), np.uint8)
    cv2.rectangle(mask, (100, 100), (w-100, h-100), 255, -1)

    # Apply bitwise operations
    and_ = cv2.bitwise_and(img, img, mask=mask)
    or_ = cv2.bitwise_or(img, img, mask=mask)
    xor_ = cv2.bitwise_xor(img, img, mask=mask)
    not_ = cv2.bitwise_not(img, mask=mask)

    titles = ["Mask", "AND", "OR", "XOR", "NOT"]
    imgs = [mask, and_, or_, xor_, not_]

    plt.figure(figsize=(20, 4))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        if i == 0:
            plt.imshow(imgs[i], cmap="gray")
        else:
            plt.imshow(cv2.cvtColor(imgs[i], cv2.COLOR_BGR2RGB))
        plt.title(titles[i])
        plt.axis("off")

    plt.tight_layout()
    plt.show()

    print("Bitwise operations completed!")
    print(f"Image size: {w}x{h} pixels")
    print(f"Mask area: {w-200}x{h-200} pixels")
else:
    print("No file uploaded.")
```

*****Task 10*****

Load an image and convert its pixel values to a Pandas DataFrame, then analyze basic statistics, also apply a mask
to it using bitwise AND operation.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import files

print("Upload an image from your PC:")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]

    img = cv2.imread(filename)

    # Create DataFrame from image pixels
    df = pd.DataFrame(img.reshape(-1, 3), columns=["B", "G", "R"])
    print("Image Pixel Statistics:")
    print("=" * 50)
    print(df.describe())
    print("=" * 50)

    # Create circular mask
    mask = np.zeros(img.shape[:2], np.uint8)
    cv2.circle(mask, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
    masked = cv2.bitwise_and(img, img, mask=mask)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(mask, cmap="gray")
    plt.title("Circular Mask")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))
    plt.title("Masked Result")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    print("Image analysis and masking completed!")
    print(f"Image dimensions: {img.shape[1]}x{img.shape[0]} pixels")
    print(f"Mask radius: 100 pixels")
else:
    print("No file uploaded.")
```
